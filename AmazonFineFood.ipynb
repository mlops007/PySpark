```
from pyspark.sql.types import StructType, StructField, TimestampType, IntegerType, StringType
from pyspark.sql.functions import from_unixtime, expr
reviews_schema = StructType([
    StructField("Id", IntegerType(), True),
    StructField("ProductId", StringType(), True),
    StructField("UserId", StringType(), True),
    StructField("ProfileName", StringType(), True),
    StructField("HelpfulnessNumerator", IntegerType(), True),
    StructField("HelpfulnessDenominator", IntegerType(), True),
    StructField("Time", IntegerType(), True),
    StructField("Summary", StringType(), True),
    StructField("Text", StringType(), True)
])
```

```
reviews_df = spark.read\
                .format("csv")\
                .option("header", False)\
                .schema(reviews_schema)\
                .load("dbfs:/mnt/finefooddata/landing/Reviews.csv", skipRows=1)
reviews_df.show()
```

```
reviews_df = reviews_df.withColumn("Time" , from_unixtime(reviews_df.Time, "yyyy-MM-dd"))
reviews_df.show()
```

```
reviews_df.printSchema()
```

```
reviews_df.rdd.getNumPartitions()
```

```
reviews_df.show()
```

```
scores_schema = StructType([
    StructField("Id", IntegerType(), True),
    StructField("Score", IntegerType(), True)
])
```

```
scores_df = spark.read\
                .format("csv")\
                .option("header", False)\
                .schema(scores_schema)\
                .load("dbfs:/mnt/finefooddata/landing/Scores.csv",skipRows=1)
scores_df.show()
```

```
scores_df.printSchema()
```

# filtering 

```
reviews_df.count()
```

```
reviews_df = reviews_df.filter(reviews_df.HelpfulnessDenominator>reviews_df.HelpfulnessNumerator)
reviews_df = reviews_df.dropDuplicates()
reviews_df = reviews_df.cache()
reviews_df.show()
```

```
reviews_df.count()
```

```
from pyspark.sql.functions import when
scores_df = scores_df.withColumn("reviews", when(scores_df.Score > 3, "Positive").when(scores_df.Score == 3, "Neutral").otherwise("Negative"))
scores_df = scores_df.drop("Score")
display(scores_df)
```

```
scores_df = scores_df.cache()
scores_df.count()
scores_df.count()
```

```
reviews_df.write\
          .format("parquet")\
          .mode("overwrite")\
          .option("path", "/mnt/finefooddata/staging/reviews_filtered")\
          .save()
```

```
scores_df.write\
          .format("parquet")\
          .mode("overwrite")\
          .option("path", "/mnt/finefooddata/staging/scores_filtered")\
          .save()
```

```
import datetime
from pyspark.sql.functions import *
from pyspark.sql.functions import year, month, dayofmonth
joined_df = joined_df.withColumn("Year", year("Time").alias("Year"))
joined_df = joined_df.withColumn("month", date_format("Time", "MMMM"))
```
