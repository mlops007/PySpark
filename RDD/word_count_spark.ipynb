```python
rdd1 = spark.sparkContext.textFile("/user/itv009033/data/input/test.txt.txt")
```

```python
rdd1.collect()
```

```python
rdd2 = rdd1.flatMap(lambda x : x.split(" "))
rdd2.collect()
```

```python
rdd3 = rdd2.map(lambda word : (word,1))
rdd3.collect()
```

```python
rdd4 = rdd3.reduceByKey(lambda x,y : (x+y))
rdd4.collect()
```
